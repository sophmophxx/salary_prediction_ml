# Sprint 0: Business Understanding und Projektstruktur

## Business Understanding

### Problemstellung und Ausgangslage
Stellenanzeigen auf Plattformen wie der Bundesagentur für Arbeit enthalten detaillierte Informationen zu Berufen, Anforderungen und Standorten. Ein entscheidendes Kriterium für Bewerbende, das Gehalt, ist jedoch häufig nicht explizit ausgewiesen. Für Jobportale und Arbeitsmarktanalysen stellt diese Intransparenz eine Hürde dar, da eine gezielte Filterung oder statistische Auswertung des Lohnniveaus erschwert wird.

### Zielsetzung des Projekts
Das Ziel dieses Projekts ist die Entwicklung eines Klassifikationsmodells, das basierend auf Merkmalen wie Jobtitel, Beruf, Region und weiteren Textinformationen ein voraussichtliches Gehaltsniveau schätzt.

**Operationalisierung:** Statt einer fehleranfälligen Regression auf exakte Euro-Werte werden drei Gehaltsklassen definiert: „niedrig“, „mittel“ und „hoch“.

**Methodischer Ansatz:** Da die Stellenanzeigen häufig selbst keine Gehaltsangaben enthalten, wird im Hauptsetup ein Proxy-Ziel verwendet, das aus externen BA-Entgeltdaten auf Ebene von Berufsgruppe und Region abgeleitet wird. Da diese Entgeltdaten für Deutschland beziehungsweise deutsche Regionen vorliegen, werden im Projekt nur Stellenanzeigen mit Arbeitsort in Deutschland berücksichtigt.

### Wirtschaftliche Relevanz
- **Nutzererfahrung:** Ermöglichung einer groben Gehaltseinordnung auch dann, wenn Arbeitgeber keine Angaben machen
- **Marktanalyse:** Automatisierte Einschätzung von Lohnstrukturen in verschiedenen Berufsgruppen und Regionen
- **Datenanreicherung:** Strukturierung unvollständiger Datensätze für weiterführende analytische Zwecke

### Warum Machine Learning?
Die Klassifizierung von Gehaltsstufen ist ein mehrdimensionales Problem, da das Gehaltsniveau mit mehreren Merkmalen einer Stellenanzeige zusammenhängt, insbesondere mit Beruf, Region und Formulierungen im Jobtitel. Machine Learning ist hier sinnvoll, weil ein Modell Muster aus diesen Merkmalen gemeinsam lernen und auf neue Anzeigen anwenden kann. Gleichzeitig wird im Projekt reflektiert, dass ein Teil der Vorhersagekraft stark aus berufs- und regionsbezogenen Informationen stammt und die Zielvariable im Hauptsetup nur als Proxy vorliegt.

### Projektplan (Sprint-Struktur)
Das Projekt folgt einem iterativen Prozess:

- **Sprint 1:** Datenakquise über die API der Bundesagentur für Arbeit, Bereinigung der Rohdaten und Exploration potenziell brauchbarer Merkmale
- **Sprint 2:** Ergänzung der KldB-Zuordnung, Verknüpfung mit externen Entgeltdaten und Konstruktion der Proxy-Zielvariable
- **Sprint 3:** Modellierung, Evaluation, Cross-Validation, Fehleranalyse und kritische Reflexion des BA-Setups
- **Sprint 4:** Prüfung einer alternativen Datenquelle mit direkteren Gehaltsfeldern über die Adzuna-API und Bewertung ihrer Datenqualität
- **Sprint 5:** Vergleich des BA- und Adzuna-Ansatzes sowie ein kleiner Machbarkeitstest mit einem direkten Adzuna-Gehaltssetup
