## Gesamtfazit

Ziel der Arbeit war es, zu untersuchen, ob sich aus Merkmalen von Stellenanzeigen ein Gehaltsniveau mit Methoden des Machine Learning vorhersagen lässt. Ausgangspunkt war dabei die Beobachtung, dass Gehaltsangaben in vielen Anzeigen nicht direkt oder nicht in ausreichend strukturierter Form vorliegen. Deshalb wurde das Projekt iterativ in mehreren Sprints aufgebaut und schrittweise weiterentwickelt.

Im ersten Teil des Projekts wurde mit Daten der Bundesagentur für Arbeit gearbeitet. Nach Datenbeschaffung, Bereinigung und Exploration wurde in Sprint 2 eine Proxy-Zielvariable konstruiert, indem Stellenanzeigen über KldB-Codes, Berufsgruppen und Regionen mit externen Entgeltdaten verknüpft wurden. Dadurch entstand eine Gehaltsklassifikation mit den Klassen niedrig, mittel und hoch. Diese Lösung war methodisch nicht ideal, da kein direkt beobachtetes individuelles Gehalt vorlag, erwies sich aber als praktikabel und ermöglichte ein stabiles überwachtes Lernproblem. Die Match-Rate lag dabei sehr hoch, sodass fast alle Anzeigen in das weitere Setup übernommen werden konnten.

In Sprint 3 wurde auf dieser Grundlage ein Klassifikationssetup aufgebaut. Die Modelle Logistic Regression und Random Forest übertrafen die Baseline deutlich. Zusätzlich zeigten Cross-Validation und Feature-Vergleich, dass die Modellleistung stabil ist und vor allem durch berufs- und regionsbezogene Informationen getragen wird. Damit konnte gezeigt werden, dass die verwendeten Merkmale ausreichen, um ein geschätztes Gehaltsniveau deutlich besser als durch eine triviale Vorhersage zu klassifizieren. Gleichzeitig wurde aber auch deutlich, dass die hohe Modellgüte vorsichtig interpretiert werden muss, da die Zielvariable selbst bereits eng mit Beruf und Region verknüpft ist.

Um diese zentrale methodische Einschränkung nicht nur zu benennen, sondern aktiv zu prüfen, wurde das Projekt in Sprint 4 und 5 erweitert. Mit der Adzuna-API wurde eine alternative Datenquelle untersucht, die direkte Gehaltsfelder enthält. Diese Quelle lag methodisch näher am eigentlichen Business Problem, erwies sich in der praktischen Nutzung jedoch als deutlich lückenhafter und stärker bereinigungsbedürftig. Nach Dublettenbereinigung und Plausibilitätsfilterung blieb nur ein vergleichsweise kleiner Datensatz übrig. Ein zusätzlicher Machbarkeitstest zeigte zwar, dass sich auch auf dieser Basis ein sinnvolles Klassifikationssetup aufbauen lässt, gleichzeitig bestätigte der Vergleich aber, dass das BA-Setup für die vorliegende Arbeit die tragfähigere Hauptlösung bleibt.

Insgesamt zeigt die Arbeit, dass sich aus den verfügbaren Daten ein konsistenter und nachvollziehbarer Data-Science-Prozess aufbauen ließ. Die Arbeit beantwortet die ursprüngliche Fragestellung daher nicht im Sinne einer exakten Gehaltsvorhersage, wohl aber im Sinne einer belastbaren Klassifikation eines geschätzten Gehaltsniveaus. Gleichzeitig macht sie deutlich, dass für eine methodisch noch stärkere Weiterentwicklung künftig Datenquellen mit direkter, vollständigerer und qualitativ stabilerer Gehaltsinformation wünschenswert wären.
