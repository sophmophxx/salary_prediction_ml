# Sprint 0: Business Understanding und Projektstruktur

## 1. Business Understanding

### 1.1 Problemstellung und Ausgangslage
Stellenanzeigen auf Plattformen wie der Bundesagentur für Arbeit enthalten detaillierte
Informationen zu Berufen, Anforderungen und Standorten.
Ein entscheidendes Kriterium für Bewerber, das **Gehalt**, ist jedoch häufig nicht explizit ausgewiesen. Für Jobportale und Arbeitsmarktanalysten stellt diese Intransparenz eine Hürde dar, da eine gezielte Filterung oder statistische Auswertung
des Lohnniveaus erschwert wird.

### 1.2 Zielsetzung des Projekts
Das Ziel dieses Projekts ist die Entwicklung eines Klassifikationsmodells, das basierend auf Merkmalen wie **Jobtitel, Region und Textinhalten** ein voraussichtliches **Gehaltsniveau** schätzt.

* **Operationalisierung:** Statt einer fehleranfälligen Regression auf exakte Euro-Werte werden drei Gehaltsklassen definiert: "niedrig", "mittel" und "hoch".
* **Methodischer Ansatz:** Da die Stellenanzeigen oft selbst keine Gehaltsangaben beinhalten, wird ein Proxy-Ziel verwendet, welches aus den Medianentgelten des **Entgeltatlasses** für den jeweiligen Beruf und die Region abgeleitet wird.

### 1.3 Wirtschaftliche Relevanz
* **Nutzererfahrung:** Ermöglichung von Gehaltsfiltern für Bewerber, selbst wenn Arbeitgeber keine Angaben machen.
* **Marktanalyse:** Automatisierte Einschätzung von Lohnstrukturen in verschiedenen Branchen und Regionen.
* **Datenanreicherung**: Strukturierung unvollständiger Datensätze für weiterführende analytische Zwecke.

### 1.4 Warum Machine Learning?
Die Klassifizierung von Gehaltsstufen ist ein komplexes Problem, da das Gehalt nicht nur vom Beruf, sondern auch von Nuancen in der Stellenbeschreibung abhängt. Machine Learning kann diese semantischen Muster erkennen, die über einfache Datenbankabfragen hinausgehen. Das Modell wird dabei gegen eine Baseline (die häufigste Gehaltsklasse) getestet, um den Mehrwert der Vorhersage zu belegen.

### 1.5 Projektplan (Sprint-Struktur)
Das Projekt folgt einem iterativen Prozess:
* **Sprint 1:** Datenakquise über die API der Bundesagentur für Arbeit und Untersuchung der Rohdaten hinsichtlich brauchbarer Features (Titel, Ort, Beschreibung).
* **Sprint 2:** Datenbereinigung mit den Proxy-Gehaltsdaten und Erstellung eines Baseline-Modells.
* **Sprint 3:** Feinabstimmung des Modells, Fehleranalyse (Confusion Matrix) und kritische Reflexion der Ergebnisse.
